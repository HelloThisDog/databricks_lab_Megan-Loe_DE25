```python
%sql
SELECT
    event_number_of_finishers,
    COUNT (DISTINCT event_id) AS event_count
FROM
    marathos_list.gold.mart_swedish_results
GROUP BY
    event_number_of_finishers
ORDER BY
    event_number_of_finishers
```

```python
%sql
SELECT
  DATE_TRUNC(
    'month',
    TO_DATE(REGEXP_EXTRACT(`event_date`, '(\\d{2}\\.\\d{2}\\.\\d{4})$', 1), 'dd.MM.yyyy')
  ) as month,
  AVG(`athlete_average_speed`) as avg_speed,
  PERCENTILE(`athlete_average_speed`, 0.5) as median_speed,
  COUNT(*) as num_results
FROM
  `marathos_list`.`gold`.`mart_swedish_results`
WHERE
  `athlete_average_speed` IS NOT NULL
  AND `event_date` IS NOT NULL
  AND `athlete_average_speed` <= 20
GROUP BY
  DATE_TRUNC(
    'month',
    TO_DATE(REGEXP_EXTRACT(`event_date`, '(\\d{2}\\.\\d{2}\\.\\d{4})$', 1), 'dd.MM.yyyy')
  )
ORDER BY
  month
```
